### Create Files for Wordcloud

```python
interimDir = "../data/interim/scraped_data/"
dirs = [folder for folder in os.listdir(interimDir) if os.path.isdir(os.path.join(interimDir, folder))]
for folder in dirs: 
    allTexts = ""
    for file in os.listdir(os.path.join(interimDir, folder)):
        with open(os.path.join(interimDir, folder, file), "r", encoding="utf-8") as f:
            text = f.read()
        allTexts += text
    with open(folder+"_allTexts.txt", "wb") as f:
        f.write(allTexts.encode("utf-8"))
```
