# Explore NER Results on Scraped Data

```python
import pickle
import os
import pandas as pd
import numpy as np
```

### Load Data

```python
with open("../data/processed/scraped_data_entitylist.pkl", "rb") as f:
    d = pickle.load(f)
```

### Examine Sample Document

```python
docNum = 120
doc = d[list(d.keys())[docNum]]
for k,v in doc.items():
    print("****", k , "****")
    print(v["sentence"])
    for e, t in zip(v["tags"], v["entities"]):
        print(e, " - ", t)
    print()    
```

### Create Files for Wordcloud

```python
interimDir = "../data/interim/scraped_data/"
dirs = [folder for folder in os.listdir(interimDir) if os.path.isdir(os.path.join(interimDir, folder))]
for folder in dirs: 
    allTexts = ""
    for file in os.listdir(os.path.join(interimDir, folder)):
        with open(os.path.join(interimDir, folder, file), "r", encoding="utf-8") as f:
            text = f.read()
        allTexts += text
    with open(folder+"_allTexts.txt", "wb") as f:
        f.write(allTexts.encode("utf-8"))
```
